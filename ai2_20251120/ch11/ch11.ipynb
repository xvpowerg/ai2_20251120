```python
def percent_missing(df):
    p_nan = (df.isnull().sum() / len(df)) * 100
    p_nan = p_nan[p_nan > 0].sort_values(ascending=False)
    return p_nan
p_nan = percent_missing(df)
p_nan
```

```python
import pandas as pd
data = pd.DataFrame({"Fruits":["Apple","Banana","Cherry","Kiwi"]})
fruit_mapping = {"Apple":1,"Banana":2,"Cherry":3,"Kiwi":4}
data["Fruit_number1"] = data["Fruits"].map(fruit_mapping)

print(data)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
# å»ºç«‹è³‡æ–™æ¡†
data = pd.DataFrame({
    'Fruits': ['Apple', 'Banana', 'Cherry', 'Kiwi','Cherry', 'Kiwi','Apple', 'Banana'],
})
le = LabelEncoder()
le.fit(data["Fruits"])
data["Fruit_number"] = le.transform(data["Fruits"])
data
```

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

data = pd.DataFrame({
    'Fruits': ['Apple', 'Banana', 'Cherry', 'Kiwi','Cherry', 'Kiwi','Apple', 'Banana'],
})

order = ['Kiwi', 'Apple', 'Banana', 'Cherry']  # ä½ æƒ³è¦çš„é †åº
enc = OrdinalEncoder(categories=[order])

data['Fruit_number'] = enc.fit_transform(data[['Fruits']]).astype(int)
data
```
