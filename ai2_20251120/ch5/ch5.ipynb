```python
def last_four(num):
    return str(num)[-4:]
```

```python
last_four(3560325168603410)
```

```python
df["CC Number"].apply(last_four)
```

```python
def yelp(price):
    if price < 10:
        return "$"
    elif price >= 10 and price < 30:
        return "$$"
    else:
        return "$$$"
yelp(20)
```

```python
df["total_bill"].mean()
```

```python
df["Expensive"] = df["total_bill"].apply(yelp)
```

```python
df.head()
```

```python
def quality(total_bill,tip):
    if tip / total_bill > 0.25:
        return "Generous"
    else:
        return "Other"
print(quality(100,15))    
```

```python
#apply é è¨­axis=0 æ‰€ä»¥é è¨­æœƒæ‰¾rowæ˜¯å¦æœ‰keyç‚º total_bill tip
func1 = lambda df:quality(df["total_bill"],df["tip"])
df["Tip Quality"] = df[["total_bill","tip"]].apply(func1,axis=1)
```
